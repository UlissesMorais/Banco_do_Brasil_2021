#### Dados de 2021 e 2022

```python
# dados brutos de 2021 e 2022
dados_2021_2022 = {}
for ano in range(2021,2023):
    print(ano)
    path = f"https://github.com/adaoduque/Brasileirao_Dataset/blob/master/data/brasileirao-{ano}.json?raw=true"
    dados_2021_2022[ano] = pd.read_json(path)
```

### Cria DataFrame vazio

```python
# colunas dataframe
colunas = ['Rodada',
           'Data',
           'Mandante',
           'Visitante',
           'Arena',
           'Mandante_Placar',
           'Visitante_Placar',
           'Mandante_Resultado']

# dataframe para receber os dados desejados
df = pd.DataFrame(data=None, columns=colunas)
```

### Preenche DataFrame com dados de 2021 e 2022

```python
# insere dados
for df_ano in dados_2021_2022.values():
    for col in df_ano.columns:
        lista = [None]*8
        for idx in df_ano.index:
            
            lista[0] = col                                      # Rodada
            lista[1] = df_ano.loc[idx, col]['date'   ]          # Data
            lista[2] = df_ano.loc[idx, col]['clubs'  ]['home']  # Mandante
            lista[3] = df_ano.loc[idx, col]['clubs'  ]['away']  # Visitante
            lista[4] = df_ano.loc[idx, col]['stadium']          # Arena
            lista[5] = df_ano.loc[idx, col]['goals'  ]['home']  # Mandante_Placar
            lista[6] = df_ano.loc[idx, col]['goals'  ]['away']  # Mandante_Placar
            
            df.loc[len(df.index)] = lista
```

### Ajusta dados de 2021 e 2022

```python
# funcao auxiliar
def sub_valor(valor):
    if valor > 0:
        return 'Vitoria'
    elif valor == 0:
        return 'Empate'
    else:
        return 'Derrota'
```

```python
# coloca gols como float e define o resultado do Mandante
df["Mandante_Placar"] = df["Mandante_Placar"].astype(np.int8)
df["Visitante_Placar"] = df["Visitante_Placar"].astype(np.int8)
df["Mandante_Resultado"] = (df["Mandante_Placar"] - df["Visitante_Placar"]).apply(lambda x: sub_valor(x))
```

### Leitura dados 2023

```python
# leitura de dados do brasileirão 2023
df_2023 = pd.read_excel('./dados/brasileirao_2023.xlsx')

# ajusta data
df_2023['Data'] = pd.to_datetime(df_2023['Data'], format="%d/%m/%Y")

# Ajusta valores
df_2023["Mandante_Placar"] = df_2023["Mandante_Placar"].astype(np.int8)
df_2023["Visitante_Placar"] = df_2023["Visitante_Placar"].astype(np.int8)

# determina o resultado para o Mandante
df_2023["Mandante_Resultado"] = (df_2023["Mandante_Placar"] - df_2023["Visitante_Placar"]).apply(lambda x: sub_valor(x))
```

### Concatena os dados

```python
df_concat = pd.concat([df_jogos, df, df_2023]).sort_values(['Data'], ascending=True, ignore_index=True)
```

```python
df_concat.tail(3)
```

```python
# padronização dos nomes dos times
lista_times = ['Athlético-PR', 'Botafogo', 'Botafogo-rj', 'CORINTHIANS', 'CRUZEIRO', 'Ceará SC', 'FLAMENGO', 'FLUMINENSE',
               'GUARANI', 'PALMEIRAS', 'PAYSANDU', 'SANTOS', 'Sport Recife', 'VASCO']

lista_correta = ['Athletico-PR', 'Botafogo-RJ', 'Botafogo-RJ', 'Corinthians', 'Cruzeiro', 'Ceará', 'Flamengo', 'Fluminense',
                 'Guarani', 'Palmeiras', 'Paysandu', 'Santos', 'Sport', 'Vasco',]

# dicionarios com as datas mapeadas
time_dict = dict(zip(lista_times, lista_correta))

# substitui as datas erradas para as datas certas
df_concat["Mandante"] = df_concat["Mandante"].replace(time_dict)
df_concat["Visitante"] = df_concat["Visitante"].replace(time_dict)
```

```python
print(df_concat.shape)
df_concat.head(3)
```

```python
# define o Vencedor e Derrotado do confronto e placar
df_concat["Vencedor"] = df_concat["Mandante_Placar"] - df_concat["Visitante_Placar"]
df_concat["Derrotado"] = df_concat["Mandante_Placar"] - df_concat["Visitante_Placar"]

# monta coluna Placar
df_concat["Placar"] = df_concat["Mandante_Placar"].apply(str) + "x" + df_concat["Visitante_Placar"].apply(str)

for idx in df_concat.index:
    
    if df_concat.loc[idx, "Vencedor"] > 0.0:
        df_concat.loc[idx, "Vencedor"] = df_concat.loc[idx, "Mandante"]
        df_concat.loc[idx, "Derrotado"] = df_concat.loc[idx, "Visitante"]
        
    elif df_concat.loc[idx, "Vencedor"] == 0.0:
        df_concat.loc[idx, "Vencedor"] = '-'
        df_concat.loc[idx, "Derrotado"] = '-'
        
    else:
        df_concat.loc[idx, "Vencedor"] = df_concat.loc[idx, "Visitante"]
        df_concat.loc[idx, "Derrotado"] = df_concat.loc[idx, "Mandante"]
```

### Inserção do período

```python
# junta os datasets e retorna apenas os registros corretos criados na junção
df_periodo['key'] = 1
df_concat['key'] = 1
 
df_final = (pd.merge(df_periodo, df_concat, on ='key')
            .query('Data >= Inicio & Data <= Fim')
            .drop(["Inicio","Fim","key"], axis=1)
            .sort_values(['Data'], ascending=True, ignore_index=True)
)

# cria um ID para cada partida
df_final["Jogo"] = "J"
df_final["Numero"] = list(map(str, list(df_final.index)))
df_final["ID"] = df_final["Jogo"] + df_final["Numero"]

# retirar colunas
df_final = df_final.drop(["Jogo", "Numero"], axis=1)

# reordenar colunas
df_final = df_final.reindex(columns=["ID"] + list(df_final.columns)[:-1])
```
